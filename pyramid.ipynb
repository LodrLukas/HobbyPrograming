# Mise: Stavba startovací rampy

Představme si, že jsme hlavním inženýrem v mezihvězdné agentuře **StarPrd**. Máme za úkol postavit startovací rampu pro novou raketu "BumajsmenaMarsu". Tato rampa musí být postavená z odolných ocelových bloků (označených symbolem `*`).

Aby raketa nespadla, rampa musí mít speciální tvar:

* Každé další patro směrem dolů musí být širší o přesně dva bloky.
* Špička rampy začíná vždy jedním blokem.
* Celková šířka základny (posledního patra) musí odpovídat plánu, který dostaneme od kapitána (vždy **liché číslo**).


### Násobení bloků (String Multiplication)

V Pythonu máme úžasný trik. Když chceme vedle sebe naskládat 10 ocelových bloků, nemusíme psát `print("**********")`.

Stačí použít násobení:

```python
bloky = "*" * 10
print(bloky)
```

Tohle funguje na jakýkoliv text. Chceš vypsat 50 mezer? Stačí `" " * 50`.
