Představme si, že máme k dispozici nekonečné plátno a poslušnou želvu, která udělá přesně to, co napíšeme do kódu. Abychom ji probudili, musíme ji nejdříve zavolat:

```python
import turtle

# Nastavení plátna a želvy
t = turtle.Turtle()
t.shape("turtle")  # Aby to vypadalo jako želva
t.speed(3)         # Nastavení rychlosti (1-10)
```

## Základní příkazy (Trénink v aréně)
Než začneme stavět složité věci, musíme se naučit želvu ovládat.

* **Pohyb:** `t.forward(100)` (dopředu) nebo `t.backward(50)` (dozadu).
* **Otáčení:** `t.right(90)` (vpravo o 90 stupňů) nebo `t.left(45)` (vlevo).
* **Barvičky:** * `t.pencolor("red")` – změní barvu "fixy".
    * `t.pensize(5)` – udělá čáru tlustší.
    * `turtle.bgcolor("black")` – změní barvu pozadí celého okna.

---

## Iterativní cykly: Geometrické obrazce

Místo toho, abychom psali `forward` a `left` pořád dokola, použijeme náš známý cyklus `for`. Tím můžeme kreslit pravidelné tvary.

**Úkol: Barevný čtverec**
```python
barvy = ["red", "blue", "green", "yellow"]

for i in range(4):
    t.pencolor(barvy[i])
    t.forward(100)
    t.left(90)
```

**Zajímavost:** Pokud změníme úhel na něco neobvyklého (třeba 170 stupňů) a necháme cyklus běžet 50x, želva vytvoří cool hvězdici!

## Mistr Geometrie (Mnohoúhelníky)

Želva je sice šikovná, ale potřebuje, abys jí prozradil tajný matematický vzorec. Aby se želva vrátila přesně tam, odkud začala (a uzavřela tvar), musí se celkem otočit o **360 stupňů**.



### Matematický trik
Pokud chceš nakreslit pravidelný n-úhelník (všechny strany stejné), stačí vydělit 360 počtem stran:
* **Čtverec (4 strany):** $360 / 4 = 90$ stupňů.
* **Trojúhelník (3 strany):** $360 / 3 = 120$ stupňů.
* **Pětiúhelník (5 stran):** $360 / 5 = 72$ stupňů.

---

## Rekurze: Želví fraktály

Pamatujete na rekurzi u Fibonacciho? V grafice ji můžeme použít k vytvoření **fraktálů** – obrazců, které se skládají z menších a menších kopií sebe sama.

### Rekurzivní strom (Fractal Tree)
Pravidlo je jednoduché: Nakresli kmen, na jeho konci se rozděl na dvě menší větve. Na konci každé z nich se znovu rozděl...



```python
def nakresli_strom(delka_vetve):
    if delka_vetve > 5:  # Zastavovací podmínka (base case)
        # Nakresli hlavní větev
        t.forward(delka_vetve)
        
        # Rekurzivní krok vpravo
        t.right(20)
        nakresli_strom(delka_vetve - 15)
        
        # Návrat a rekurzivní krok vlevo
        t.left(40)
        nakresli_strom(delka_vetve - 15)
        
        # Návrat do původní pozice
        t.right(20)
        t.backward(delka_vetve)

# Spuštění
t.left(90) # Otočíme želvu nahoru, aby strom rostl vzhůru
t.pencolor("brown")
nakresli_strom(75)
```

## Co dál? Želví "L-Systémy" a umění

Napadlo by tě, že želvu můžeš ovládat i pomocí náhody? Tomu se říká **generativní umění**.

**Úkol: Náhodná procházka (Random Walk)**
Želva se v každém kroku rozhodne pro náhodnou barvu a náhodný směr. Tím vznikne unikátní "mapa" tvého algoritmu.

```python
import random

for _ in range(200):
    t.pencolor(random.random(), random.random(), random.random()) # Náhodná RGB barva
    t.forward(30)
    t.setheading(random.choice([0, 90, 180, 270])) # Náhodný směr (jen do pravých úhlů)
```
