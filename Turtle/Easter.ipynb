# Velikonoční vajíčkobraní

Zajíček rozházel po zahradě vajíčka různých barev. Tvým úkolem je vytvořit systém pro "vajíčkovou statistiku". Musíš zjistit, kolik vajíček od každé barvy děti našly, a uložit to do souboru pro babičku, aby věděla, kolik má připravit odměn.



### 1. Krok: Seznam nálezů
Představ si, že máš soubor `nalezy.txt`, kde jsou pod sebou napsané barvy vajíček tak, jak je děti postupně nosily do košíku.

**Příklad obsahu `nalezy.txt`:**
```text
červená
modrá
červená
žlutá
zelená
modrá
červená
```

---

### Úkol: Vajíčkový sčítač (Counter)

Napiš program, který:
1. Přečte barvy ze souboru `nalezy.txt`.
2. Spočítá, kolik vajíček každé barvy v košíku je (použij k tomu **slovník**).
3. Zapíše výslednou tabulku do souboru `statistiky.txt`.

**Šablona pro tebe:**


````python
# 1. Načtení dat
nalezy = []
with open("nalezy.txt", "r", encoding="utf-8") as f:
    for radek in f:
        barva = radek.strip().lower() # Odstraní mezery a sjednotí na malá písmena
        if barva: # Přeskočí prázdné řádky
            nalezy.append(barva)

# 2. Sčítání pomocí slovníku
# Klíč bude "barva" a hodnota bude "počet"
kosik = {}

for barva in nalezy:
    if barva in kosik:
        kosik[barva] += 1
    else:
        kosik[barva] = 1

# 3. Zápis statistik
with open("statistiky.txt", "w", encoding="utf-8") as f:
    f.write("--- VELIKONOČNÍ STATISTIKA ---\n")
    for barva, pocet in kosik.items():
        f.write(f"{barva.capitalize()}: {pocet} ks\n")

print("Statistika byla uložena do souboru!")
````
