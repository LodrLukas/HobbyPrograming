### Jak na to v praxi?
V programování k tomu používáme:

- Frontu (Queue): Seznam míst, která jsme viděli, ale ještě jsme je neprozkoumali (říkáme jim Open - otevřené nody).

- Množinu navštívených (Visited): Místa, kde už barva byla, abychom se netočili v kruzích (Closed - uzavřené nody).

### Mapa bludiště
Bludiště si uložíme jako 2D pole (seznam seznamů).

- \# = Skála (zeď)

- . = Volná cesta

- S = Start (kde je robot)

- E = Exit (východ)

### 1. Úkol: Naprogramuj mozek robota
Tvým úkolem je doplnit hlavní smyčku algoritmu. Robot se musí podívat na všechny sousedy (nahoru, dolů, doleva, doprava), a pokud tam není zeď a ještě tam nebyl, přidá si je do fronty.



````python
import collections
import time
import os

def vykresli_bludiste(mapa, fronta, navstiveno, aktualni):
    # Vyčistí terminál (aby to vypadalo jako animace)
    os.system('cls' if os.name == 'nt' else 'clear')
    
    for r, radek in enumerate(mapa):
        radek_text = ""
        for c, bunka in enumerate(radek):
            if (r, c) == aktualni:
                radek_text += "S " # Aktuální pozice robota
            elif (r, c) in fronta:
                radek_text += "? "  # Open - čeká ve frontě
            elif (r, c) in navstiveno:
                radek_text += ". "  # Closed - už jsme prozkoumali
            else:
                radek_text += bunka + " "
        print(radek_text)
    print("\nLegenda: S robot, ? v pořadníku, . prozkoumáno, # zeď")
    time.sleep(0.3) # Zpomalení, abychom viděli, co se děje

def bfs_utek(mapa, start):
    fronta = collections.deque([start])
    navstiveno = {start}
    
    while fronta:
        aktualni = fronta.popleft() # Vezmeme prvního v řadě
        r, c = aktualni
        
        # 1. Zkontroluj, jestli jsme u východu 'E'
        if mapa[r][c] == 'E':
            print("VÍTĚZSTVÍ! Robot našel cestu ven!")
            return True
            
        # 2. Prozkoumej sousedy (nahoru, dolů, vlevo, vpravo)
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nr, nc = r + dr, c + dc
            
            # TODO: Podmínka
            # Pokud je (nr, nc) v mapě, není to zeď '#' a ještě není v 'navstiveno':
            if 0 <= nr < len(mapa) and 0 <= nc < len(mapa[0]):
                if mapa[nr][nc] != '#' and (nr, nc) not in navstiveno:
                    navstiveno.add((nr, nc))
                    fronta.append((nr, nc))
        
        # Vizualizace kroku
        vykresli_bludiste(mapa, fronta, navstiveno, aktualni)
        
    print("Robot uvízl... cesta neexistuje.")
    return False

# --- MAPA PODZEMÍ ---
labyrint = [
    ["#", "#", "#", "#", "#", "#", "#", "#", "#"],
    ["#", "S", ".", ".", ".", "#", ".", ".", "#"],
    ["#", "#", "#", ".", "#", "#", ".", "#", "#"],
    ["#", ".", ".", ".", ".", ".", ".", ".", "#"],
    ["#", ".", "#", "#", "#", "#", "#", ".", "#"],
    ["#", ".", ".", ".", ".", ".", "#", "E", "#"],
    ["#", "#", "#", "#", "#", "#", "#", "#", "#"],
]

bfs_utek(labyrint, (1, 1))
````
