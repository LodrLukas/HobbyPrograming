### 1. Motivace (Příběh)
Dosud jsme pro hrdinu potřebovali spoustu proměnných: `jmeno_hrdiny`, `zdravi_hrdiny`, `sila_hrdiny`. Co když ale budeme chtít mít v naší hře 100 hrdinů? Budeme psát 300 proměnných? **Ani náhodou.**

Vytvoříme si **Třídu (Class)**. To je jako "stavební plánek" nebo "občanský průkaz", který říká, co každý hrdina musí mít, aby mohl existovat.



---

### 2. Anatomie Třídy (Vysvětlení pojmů)
Je to jako stavět lego...
* **Class (Třída):** Návod v krabici Lega. Říká ti, jak postavit auto.
* **Instance / Object (Objekt):** To konkrétní auto, které jsi právě postavil na stole.
* **Attributes (Atributy):** Barva kostiček, počet kol, jméno řidiče.

```python
class Hrdina:
    # Magická funkce, která se spustí, když hrdina "vniká" do světa
    def __init__(self, jmeno, povolani, zdravi):
        self.jmeno = jmeno        # Atribut: Jak se jmenuje?
        self.povolani = povolani  # Atribut: Co dělá?
        self.zdravi = zdravi      # Atribut: Kolik vydrží?




# Tady voláme náš plánek a dáváme mu konkrétní data
valecnik = Hrdina("Aragon", "Bojovník", 100)
carodej = Hrdina("Gandalf", "Mág", 80)

# K vlastnostem se dostaneme pomocí tečky .
print(f"Vítej, {valecnik.jmeno}! Tvé povolání je {valecnik.povolani}.")
print(f"{carodej.jmeno} má na začátku {carodej.zdravi} HP.")
```

> **Poznámka:** To divné slovo `self` je jako prst, kterým hrdina ukazuje na sebe. Říká: *"Moje (self) jméno je toto."*
