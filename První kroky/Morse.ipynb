### Jak mluvit se soubory
V Pythonu používáme pro práci se soubory příkaz `with open()`. Je to jako otevřít knihu – když ji dočteš, musíš ji zase zavřít (což `with` udělá automaticky za tebe).

**Čtení ze souboru:**


```python
# 'w' znamená write (psát) - POZOR: smaže starý obsah souboru!
with open("morseovka.txt", "w", encoding="utf-8") as soubor:
    soubor.write(".... --- .---") # Toto zapíše "AHOY"


# 'r' znamená read (číst)
with open("zprava.txt", "r", encoding="utf-8") as soubor:   
    obsah = soubor.read()
    print(f"V souboru je napsáno: {obsah}")
```

### Morseův slovník
Aby program věděl, co má překládat, připravil jsem pro tebe "překladovou tabulku" (slovník neboli `dict`).


```python
morseova_abeceda = {
    'A': '.-', 'B': '-...', 'C': '-.-.', 'D': '-..', 'E': '.', 
    'F': '..-.', 'G': '--.', 'H': '....', 'CH': '----', 'I': '..', 
    'J': '.---', 'K': '-.-', 'L': '.-..', 'M': '--', 'N': '-.', 
    'O': '---', 'P': '.--.', 'Q': '--.-', 'R': '.-.', 'S': '...', 
    'T': '-', 'U': '..-', 'V': '...-', 'W': '.--', 'X': '-..-', 
    'Y': '-.--', 'Z': '--..', ' ': '/'
}
```

### Úkol: Automatický překladač

Napiš program, který:
1.  Přečte text ze souboru `vstup.txt` (vytvoř si ho v editoru a napiš tam třeba "AHOJ").
2.  Převede každé písmenko na kód z Morseovy abecedy.
3.  Výsledek uloží do nového souboru `vystup.txt`.

**Šablona pro tebe:**


```python
# 1. Načti zprávu
with open("vstup.txt", "r", encoding="utf-8") as f:
    text = f.read().upper() # .upper() zajistí, že všechna písmena budou velká

prelozeny_text = ""

# 2. Projdi písmeno po písmenu a přelož ho
for pismeno in text:
    if pismeno in morseova_abeceda:
        prelozeny_text += morseova_abeceda[pismeno] + " "

# 3. Ulož výsledek
with open("vystup.txt", "w", encoding="utf-8") as f:
    f.write(prelozeny_text)

print("Tajná zpráva je v souboru vystup.txt")
```
