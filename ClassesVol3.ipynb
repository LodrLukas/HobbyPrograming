### Profesionální struktura složek



Aby se nám v projektu dobře orientovalo, nebudeme mít všechny soubory na jedné hromadě. Vytvoříme si pro naše "součástky" speciální složku. Takhle to vypadá v opravdovém světě:


*   📂 **MojeVesmirnaHra/** (Hlavní složka projektu)

### Implementace do naší hry (Dělení na moduly)

Teď tu naši velkou hru s RPG hrdiny a soubory rozstříháme a uklidíme do šuplíků.

#### 📁 Šuplík 1: `soucastky/nepratele.py`
Tady jen definujeme, jak vypadají a co umí. Žádná akce, jen plány.


```python
class Potvora:
    def __init__(self, jmeno, zdravi):
        self.jmeno = jmeno
        self.zdravi = zdravi

class Alien(Potvora):
    def utok(self):
        return f"{self.jmeno} vystřelil sliz!"
```

#### 📁 Šuplík 2: `soucastky/zapisnik.py`
Tady máme funkce pro práci se soubory (převádění čísel, ukládání historie).


```python
def uloz_bitvu(vysledek):
    with open("denik_hrace.txt", "a", encoding="utf-8") as f:
        f.write(vysledek + "\n")
    print("Zápis do deníku dokončen.")
```

#### Hlavní mozek: `main.py`
Tady všechno propojíme. (Výhody: krátké, přehledné, prostě žůžo)

Místo 200 řádků kódu to je třeba 10,



```python
# Importujeme konkrétní věci z našich součástek
from soucastky.nepratele import Alien
from soucastky.zapisnik import uloz_bitvu

# 1. Oživíme potvoru
emzak = Alien("Zorg", 100)

# 2. Akce
vypis = emzak.utok()
print(vypis)

# 3. Uložíme si to na disk
uloz_bitvu(f"Bitva s {emzak.jmeno}: {vypis}")
```


### Shrnutí:
*   **Modul:** Je jeden soubor s kódem.
*   **Balíček:** Je složka, která obsahuje víc modulů (souborů).
*   **Import:** Je most, po kterém přecházejí třídy a funkce z jednoho souboru do druhého.
