### Extra náročná úloha: Automatický Anti-Spam filtr
Server čelí útoku botů! Musíš napsat pokročilý filtr, který do hry nepustí nikoho, kdo vypadá podezřele.

Pravidla bezpečnosti pro dnešní den:

- Žádný ban: Hráč nesmí být v seznamu ban_list.

- Délka jména: Nick musí mít alespoň 3 znaky (příliš krátké nicky jsou podezřelé).

- Pravidlo podtržítka: Administrátor nemá rád jména, která začínají podtržítkem _ (často to dělají boti).

- Maximální délka: Jméno nesmí být delší než 15 znaků (nevejde se do tabulky skóre).

Tvůj úkol: Napiš funkci anti_spam_filtr(pripojovani, zabanovani), která vrátí seznam jmen, která splňují všechna pravidla výše.

```python
def anti_spam_filtr(pripojovani, zabanovani):
    finalni_seznam = []
    
    for hrac in pripojovani:
        # TADY DOPLŇ PODMÍNKY:
        # 1. hrac není v zabanovani
        # 2. délka len(hrac) je mezi 3 a 15 znaky
        # 3. hrac[0] není "_"
        continue
            
    return finalni_seznam

# Zkušební data:
pokusy_o_pripojeni = ["_Bot1", "Ed", "SuperGamer2026", "Hacker_Joe", "Karel", "TohleJeMocDlouheJmenoProServer"]
zabanovani = ["Hacker_Joe"]

vysledek = anti_spam_filtr(pokusy_o_pripojeni, zabanovani)
print(f"Hráči, kteří prošli filtrem: {vysledek}")
# Očekávaný výsledek: ["SuperGamer2026", "Karel"]
```
