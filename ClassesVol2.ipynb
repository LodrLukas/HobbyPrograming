# Metody a interakce
### Rozdíl mezi funkcí a metodou
Dosud jsme psali funkce jako `nakresli_ctverec()`. Metoda je skoro to samé, ale patří konkrétnímu hrdinovi. 

**Příklad:** 
* **Funkce:** `jist()` – obecná činnost.
* **Metoda:** `aragon.jist()` – konkrétní Aragon se právě najedl a jemu se zvedne zdraví.


---

### Kouzelné slovíčko `self` (Znovu a lépe)
##### Proč musí být `self` v každé závorce?
> Když učitel řekne ve třídě: *"Otevři si sešit,"* každý žák ví, že si má otevřít **svůj** sešit. `self` je pro hrdinu ten "můj sešit". Bez něj by hrdina nevěděl, čí zdraví má vlastně měnit.



---

### Hrdina v akci
Pojďme naši třídu z minulého týdne vylepšit o další schopnosti.


```python
class Hrdina:
    def __init__(self, jmeno, zdravi, sila):
        self.jmeno = jmeno
        self.zdravi = zdravi
        self.sila = sila

    # METODA: Hrdina se představí
    def pozdrav(self):
        print(f"Zdravím tě! Já jsem {self.jmeno} a mám {self.zdravi} životů.")

    # METODA: Hrdina se vyléčí
    def vypij_lektvar(self):
        self.zdravi += 20
        print(f"{self.jmeno} vypil lektvar a má nyní {self.zdravi} HP.")

    # METODA: Interakce (Útok na jiného hrdinu)
    def utoc_na(self, nepritel):
        print(f"{self.jmeno} útočí na {nepritel.jmeno}!")
        nepritel.zdravi -= self.sila
        print(f"{nepritel.jmeno} má nyní jen {nepritel.zdravi} HP.")
```


---

### "Souboj hrdinů"


```python
# Vytvoření hrdinů
aragon = Hrdina("Aragon", 100, 15)
gimli = Hrdina("Gimli", 120, 12)

# Akce
aragon.pozdrav()
gimli.pozdrav()

# Souboj!
aragon.utoc_na(gimli)
gimli.vypij_lektvar()
gimli.utoc_na(aragon)
```


---

### Proč je to důležité? (Aha! moment)
Elegance:  nemusíme psát:
`zdravi_gimliho = zdravi_gimliho - sila_aragona`

Stačí napsat:
`aragon.utoc_na(gimli)`

Program teď vypadá skoro jako příběh. Je to přehledné a dává to smysl.
