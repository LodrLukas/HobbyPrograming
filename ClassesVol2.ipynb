## Dědičnost (Inheritance) – "Jablko nepadá daleko od stromu"

**Teorie:** Programátoři jsou **líní**. Nechtějí psát stejný kód desetkrát.

Představ si, že děláš hru jako Minecraft. Máš tam `Zombíka`, `Kostlivce` a `Creepera`. 

Všichni mají zdraví, všichni se hýbou a všichni mohou dostat ránu. Místo abychom to psali pro každého zvlášť, 

vytvoříme jednoho "Pra-Moba" (Rodiče) a ostatní od něj tyhle vlastnosti **zdědí**.

---


### 1. Společný základ (Třída Rodič)
Nejdříve vytvoříme obecnou šablonu pro všechny příšery v naší hře.


```python
class Nepritel:
    def __init__(self, jmeno, zdravi):
        self.jmeno = jmeno
        self.zdravi = zdravi

    def pohni_se(self):
        print(f"{self.jmeno} se sune směrem k tobě!")

    def uber_zdravi(self, kolik):
        self.zdravi -= kolik
        print(f"{self.jmeno} dostal zásah! Zbývá mu {self.zdravi} HP.")
```


---

### 2. Specializace (Třídy Potomci)
Teď vytvoříme konkrétní příšery. Všimni si té závorky `(Nepritel)` – ta říká Pythonu, že tato nová třída má umět všechno, co umí `Nepritel`.


```python
class Zombie(Nepritel):
    def kousni(self):
        print(f"{self.jmeno} tě kousnul! Fuj, teď jsi zelený.")

# Creeper je Nepřítel, ale navíc vybuchuje
class Creeper(Nepritel):
    def vybuchni(self):
        print(f"--- BUM! --- {self.jmeno} zničil všechno kolem sebe!")
        self.zdravi = 0 # Creeper při výbuchu zmizí
```

### 3. "Armáda příšer"

```python
zora = Zombie("Zora", 50)
pepa = Creeper("Pepa", 20)

zora.pohni_se()   # Metoda od rodiče
zora.kousni()     # Vlastní metoda zombíka

pepa.pohni_se()   # Metoda od rodiče
pepa.vybuchni()   # Vlastní metoda creepera
```


---

### 4. Přepisování metod (Overriding) – "Já to dělám jinak!"
Tohle je důležitý koncept. Potomek může metodu od rodiče zdědit, ale **upravit si ji**.

**Příklad:** Každý nepřítel se hýbe, ale `Duch` by se měl hýbat jinak než `Zombík`.


```python
class Duch(Nepritel):
    def pohni_se(self):
        print(f"{self.jmeno} se nehlučně vznáší skrze zeď...")
```


Když teď zavoláš `duch.pohni_se()`, nespustí se ta nudná verze od `Nepritel`, ale tato nová "duchařská".
